```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Carregar o arquivo de áudio (substitua pelo caminho do arquivo de áudio)
arquivo_audio = "media\Do.wav"
```

```python
y, sr = librosa.load(arquivo_audio)
```

```python
plt.figure(figsize=(14, 5))
```

```python
librosa.display.waveshow(y, sr=sr)
plt.title("Forma de onda do áudio")
plt.show()
```

```python
# Usar a Transformada de Fourier para converter o som em frequências
fft = np.fft.fft(y)
```

```python
# Frequências dominantes
magnitude = np.abs(fft)
frequency = np.linspace(0, sr, len(magnitude))
```

```python
# Plotando o espectro de frequência
plt.figure(figsize=(14, 5))
plt.plot(frequency[:len(frequency)//2], magnitude[:len(magnitude)//2])
plt.title("Espectro de Frequência")
plt.xlabel("Frequência (Hz)")
plt.ylabel("Magnitude")
plt.show()
```

```python
# Dicionário de notas musicais e suas frequências padrão
notas_frequencias = {
    'C4': 261.63, 'C#4': 277.18, 'D4': 293.66, 'D#4': 311.13, 'E4': 329.63, 'F4': 349.23,
    'F#4': 369.99, 'G4': 392.00, 'G#4': 415.30, 'A4': 440.00, 'A#4': 466.16, 'B4': 493.88
}
```

```python
def freq_para_nota(frequencia):
    """Função para mapear uma frequência para uma nota musical."""
    min_diff = float('inf')
    nota_mais_proxima = None
    
    for nota, freq in notas_frequencias.items():
        diff = abs(frequencia - freq)
        if diff < min_diff:
            min_diff = diff
            nota_mais_proxima = nota
            
    return nota_mais_proxima
```

```python
# Teste: Pegue a frequência mais forte e encontre a nota musical correspondente
indice_frequencia_dominante = np.argmax(magnitude)
frequencia_dominante = frequency[indice_frequencia_dominante]

nota_detectada = freq_para_nota(frequencia_dominante)
print(f"A nota detectada é: {nota_detectada}")
```
